```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/ThaWeatherman/scrapers/master/boardgamegeek/games.csv"
games = pd.read_csv(url)
games.head()
```

```python
games.describe()
```

```python
y = games["average_rating"]
y.hist(bins = 10)
plt.show()
```

```python
print(games[games['average_rating'] == 0].iloc[0]) # first game that has average rating = 0
```

```python
print(games[games['average_rating'] > 0].iloc[0]) #first game that has average rating > 0 
```

```python
# shows that main difference between games with zero rating and more than zero rating is number of users that have reviewed the game -- users_rated
# so lets remove the games that have 0 user reviews and with missing values

games = games[games['users_rated'] > 0]
games = games.dropna(axis = 0)
games.describe()
```

```python
y = games['average_rating']
y.hist(bins = 10)
plt.show()
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters = 5, random_state = 1)
cols = games._get_numeric_data()
km.fit(cols)
labels = km.labels_

#too many columns so our plot will have tooooooo many dimensions, but we cant visualize anything w more than 3 dimensions, 
#so we gotta compress our columns using PCA (principal component analysis)
```

```python
from sklearn.decomposition import PCA

pca = PCA(2)

plotcols = pca.fit_transform(cols)
plt.scatter(plotcols[:, 0], plotcols[:, 1], c = labels)
plt.show()
```

```python
#lets figure out what features actually impact average rating

games.corr()['average_rating']
```

```python
#id and average weight have the highest correlation
#higher the id maybe more recent the game and so better reviewed
#average weight is complexity of game so maybe more complex games are better reviewed

#bayes_average_rating is derived from average rating, so we gotta get rid of it

columns = games.columns.tolist()
columns = [c for c in columns if c not in ["bayes_average_rating", "average_rating", "type", "name"]]

target = "average_rating"

columns
```

```python
from sklearn.model_selection import train_test_split as tts
x = games[columns]
y = games["average_rating"]
x_train, x_val, y_train, y_val = tts(x, y)
print(x_train.shape)
print(x_val.shape)
print(y_train.shape)
print(y_val.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error as mse

preds = lr.predict(x_val)
print(mse(preds, y_val))
#high mse so lets try a diff model

preds[0:5]
```

```python
y_val[0:5]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import model_selection
from sklearn.svm import SVC
models = []
models.append(('LR', LinearRegression()))
models.append(('RF', RandomForestRegressor()))
#models.append(('KNN', KNeighborsClassifier()))
#models.append(('DTC', DecisionTreeClassifier()))
#models.append(('G', GaussianNB()))
#models.append(('SVM', SVC()))
# evaluate each model in turn
results = []
names = []
for name, model in models:
    model.fit(x_train, y_train)
    preds = model.predict(x_val)
    mse_results = mse(preds, y_val)
    results.append(mse_results)
    names.append(name)
    msg = "%s: %f " % (name, mse_results)
    print(msg)
```
